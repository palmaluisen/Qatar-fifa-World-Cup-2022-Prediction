```python
#Load libraries
import pandas as pd 
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
#Read data
wcm = pd.read_csv('data/world_cup_matches.csv')
im = pd.read_csv('data/international_matches.csv')
#wc = pd.read_csv('data/world_cups.csv')
ranking=pd.read_csv('data/2022_world_cup_groups.csv')
qatar_matches=pd.read_csv('data/2022_world_cup_matches.csv')
```

## Construcción de la clase Pipeline_Fifa() para cargar, limpiar, ordenar y crear dataframes para entrenar el modelo de clasificación y predicción de partidos del Mundial Qatar 2022

```python
class Pipeline_Fifa:
    def __init__(self): #creación de un constructor vacio
        pass

    #Definir un método de preprocesamiento de los datos
    def Load_Data(self, wcm, im):
        wcm.rename(lambda x: x.lower().strip().replace(' ', '_'), axis='columns', inplace=True)
        im.rename(lambda x: x.lower().strip().replace(' ', '_'), axis='columns', inplace=True)
        #wc.rename(lambda x: x.lower().strip().replace(' ', '_'), axis='columns', inplace=True) 

        self.wcm                =   wcm
        self.im                 =   im
        #self.wc                 =   wc

    def Create_Dfs(self):
        ##############################
        ###Creación de df "matches"### 
        ##############################

        #En donde se concatenan los partidos de mundiales y los internacionales
        #a partir del año 1930 (por ser este el primer mundial)

        col                     =   ['date','home_team','home_goals','away_goals','away_team']
        im_df                   =   self.im[col]
        #Columna 'wcm' es variable binaria [1: si es partido de un mundial, 0: sino]
        im_df['wcm']            =   0
        wcm_df                  =   self.wcm[col]
        wcm_df['wcm']           =   1

        matches                 =   pd.concat([im_df, wcm_df])
        matches                 =   matches.reset_index()

        #Bucle para cambiar de fecha a año 
        for i in range(len(matches)):
            matches.loc[i,'year']=matches.date[i][:4] #solo coger los primeros 4 caracteres
        
        matches['year']         =   matches['year'].astype(int)
        matches                 =   matches.drop(['index','date'], axis=1)
        matches                 =   matches[matches['year']>=1930]
        matches                 =   matches.reset_index()

        #Crear una nueva variable 'status' que indica si el equipo de casa ha ganado, perdido o empatado
        #1 si ganó
        #2 si empató
        #3 si perdió

        match_result_condition  =   [
                                    (matches['home_goals']-matches['away_goals'] > 0),
                                    (matches['home_goals']-matches['away_goals'] < 0),
                                    (matches['home_goals']-matches['away_goals'] == 0)]
        
        match_result            =   [1, 3, 2] 
        matches['status']       =   np.select(match_result_condition, match_result, default='Not Specified')
        #A matches se le agregará posteriormente las columnas de atributos de los equipos

        ################################
        ###Creación de df "countries"### 
        ################################

        #En el df 'countries' contiene los atributos de cada selección, los atributos para este modelo serán:
        #la cantidad de partidos jugados, ganados, empatados y perdidos, goles anotados y recibidos tanto en 
        #mundiales como en encuentros internacionales. 

        #Crear el index del df 'countries', el cual consiste en la unión de los equipos que están como equipo
        #de casa y visitante (home_team & away_team) en mundiales y en encuentros internacionales
        xim                     =   self.im['home_team']
        yim                     =   self.im['away_team']
        countries_im            =   list(np.union1d(xim, yim))

        xwcm                    =   wcm['home_team']
        ywcm                    =   wcm['away_team']
        countries_wcm           =   list(np.union1d(xwcm, ywcm))

        countries_l             =   list(np.union1d(countries_im, countries_wcm))
        countries               =   pd.DataFrame(countries_l)
        countries.index         =   countries_l
        countries               =   countries.drop(0, axis=1)

        #Agregar los partidos de encuentros internacionales
        for i in list(countries.index):
            #partidos jugados
            countries.loc[i,'inter_match']=len(self.im[(self.im['home_team']==i)|(self.im['away_team']==i)])
            #partidos ganados
            countries.loc[i,'inter_match_won']=len(self.im[((self.im['home_team']==i)&(self.im['home_goals']>self.im['away_goals']))|((self.im['away_team']==i)&(self.im['home_goals']<self.im['away_goals']))])
            #partidos perdidos
            countries.loc[i,'inter_match_lost']=len(self.im[((self.im['home_team']==i)&(self.im['home_goals']<self.im['away_goals']))|((self.im['away_team']==i)&(self.im['home_goals']>self.im['away_goals']))])
            #partidos empatados
            countries.loc[i,'inter_match_tied']=len(self.im[((self.im['home_team']==i)&(self.im['home_goals']==self.im['away_goals']))|((self.im['away_team']==i)&(self.im['home_goals']==self.im['away_goals']))])

            #Goles anotados y recibidos
            #goles anotados
            countries.loc[i,'inter_goals_sco']=self.im.loc[self.im['home_team']==i,'home_goals'].sum() + self.im.loc[self.im['away_team']==i,'away_goals'].sum()
            #goles recibidos
            countries.loc[i,'inter_goals_con']=self.im.loc[self.im['home_team']==i,'away_goals'].sum() + self.im.loc[self.im['away_team']==i,'home_goals'].sum()

        #Agregar los partidos de mundiales anteriores 
        for i in list(countries.index):
            #partidos jugados en mundiales
            countries.loc[i,'wc_match']=len(self.wcm[(self.wcm['home_team']==i)|(self.wcm['away_team']==i)])
            #partidos ganados
            countries.loc[i,'wc_match_won']=len(self.wcm[((self.wcm['home_team']==i)&(self.wcm['home_goals']>self.wcm['away_goals']))|((self.wcm['away_team']==i)&(self.wcm['home_goals']<self.wcm['away_goals']))])
            #partidos perdidos
            countries.loc[i,'wc_match_lost']=len(self.wcm[((self.wcm['home_team']==i)&(self.wcm['home_goals']<self.wcm['away_goals']))|((self.wcm['away_team']==i)&(self.wcm['home_goals']>self.wcm['away_goals']))])
            #partidos empatados
            countries.loc[i,'wc_match_tied']=len(self.wcm[((self.wcm['home_team']==i)&(self.wcm['home_goals']==self.wcm['away_goals']))|((self.wcm['away_team']==i)&(self.wcm['home_goals']==self.wcm['away_goals']))])

            #Goles anotados y recibidos en mundiales
            #goles anotados
            countries.loc[i,'wc_goals_sco']=self.wcm.loc[self.wcm['home_team']==i,'home_goals'].sum() + self.wcm.loc[self.wcm['away_team']==i,'away_goals'].sum()
            #goles recibidos
            countries.loc[i,'wc_goals_con']=self.wcm.loc[self.wcm['home_team']==i,'away_goals'].sum() + self.wcm.loc[self.wcm['away_team']==i,'home_goals'].sum()

        #En el df matches agregar diferencias de atributos de los equipos
        for i in range(len(matches)):
            matches.loc[i,'dif_inter_match']=countries.loc[matches.loc[i,'home_team'],'inter_match'] - \
            countries.loc[matches.loc[i,'away_team'],'inter_match']

            matches.loc[i,'dif_inter_match_won']=countries.loc[matches.loc[i,'home_team'],'inter_match_won'] - \
            countries.loc[matches.loc[i,'away_team'],'inter_match_won']

            matches.loc[i,'dif_inter_match_lost']=countries.loc[matches.loc[i,'home_team'],'inter_match_lost'] - \
            countries.loc[matches.loc[i,'away_team'],'inter_match_lost']

            matches.loc[i,'dif_inter_match_tie']=countries.loc[matches.loc[i,'home_team'],'inter_match_tied'] - \
            countries.loc[matches.loc[i,'away_team'],'inter_match_tied']

            matches.loc[i,'dif_inter_goals_sco']=countries.loc[matches.loc[i,'home_team'],'inter_goals_sco'] - \
            countries.loc[matches.loc[i,'away_team'],'inter_goals_sco']

            matches.loc[i,'dif_inter_goals_con']=countries.loc[matches.loc[i,'home_team'],'inter_goals_con'] - \
            countries.loc[matches.loc[i,'away_team'],'inter_goals_con']

            matches.loc[i,'dif_wc_match']=countries.loc[matches.loc[i,'home_team'],'wc_match'] - \
            countries.loc[matches.loc[i,'away_team'],'wc_match']

            matches.loc[i,'dif_wc_match_won']=countries.loc[matches.loc[i,'home_team'],'wc_match_won'] - \
            countries.loc[matches.loc[i,'away_team'],'wc_match_won']

            matches.loc[i,'dif_wc_match_lost']=countries.loc[matches.loc[i,'home_team'],'wc_match_lost'] - \
            countries.loc[matches.loc[i,'away_team'],'wc_match_lost']

            matches.loc[i,'dif_wc_match_tied']=countries.loc[matches.loc[i,'home_team'],'wc_match_tied'] - \
            countries.loc[matches.loc[i,'away_team'],'wc_match_tied']

            matches.loc[i,'dif_wc_goals_sco']=countries.loc[matches.loc[i,'home_team'],'wc_goals_sco'] - \
            countries.loc[matches.loc[i,'away_team'],'wc_goals_sco']

            matches.loc[i,'dif_wc_goals_con']=countries.loc[matches.loc[i,'home_team'],'wc_goals_con'] - \
            countries.loc[matches.loc[i,'away_team'],'wc_goals_con']

            self.matches        =   matches[['home_team', 'home_goals', 'away_goals', 'away_team', 'wcm',
                                             'dif_inter_match', 'dif_inter_match_won', 'dif_inter_match_lost', 'dif_inter_match_tie',
                                             'dif_inter_goals_sco', 'dif_inter_goals_con',
                                             'dif_wc_match', 'dif_wc_match_won', 'dif_wc_match_lost', 'dif_wc_match_tied',
                                             'dif_wc_goals_sco', 'dif_wc_goals_con','status']]

            self.countries      =   countries

    def PCA_Data_Exploration (self, df=pd.DataFrame(), n_components=2):
        self.n_components=n_components
        self.y=self.matches.columns[-1]
        
        #Construcción de pcaDf
        if df.empty:
            df=self.matches.drop(['home_team','home_goals', 'away_goals','away_team'], axis=1)
        else:
            try: #Si el df que se introduce no tiene estas variables generará un error al intentar eliminarlas
                df=df.drop(['home_team','home_goals', 'away_goals','away_team'], axis=1)
            except:
                df=df.copy()
        
        status_ ={'1':'Victoria', '2':'Empate', '3':'Pérdida'}
        df[self.y] = df[self.y].map(status_)
        
        #La metodología de PCA se hizo con ayuda de la siguiente fuente: 
        #https://blog.damavis.com/analisis-de-componentes-principales-implementacion-en-python/
        
        #Estandarización
        X=df.drop(self.y, axis=1)

        from sklearn.preprocessing import StandardScaler
        scaler = StandardScaler()
        X = scaler.fit_transform(X)

        #PCA
        from sklearn.decomposition import PCA
        PCA = PCA(n_components=self.n_components)
        components = PCA.fit_transform(X)

        #Varianza explicada
        cumVar = pd.DataFrame(np.cumsum(PCA.explained_variance_ratio_)*100, columns=["cumVarPerc"])
        expVar = pd.DataFrame(PCA.explained_variance_ratio_*100, columns=["VarPerc"])
        var_Df=pd.concat([expVar, cumVar], axis=1)

        for i in range(self.n_components):
            var_Df=var_Df.rename(index={i: "PC"+str(i+1)})
        
        self.var_explained=var_Df

        componentsDf = pd.DataFrame(data = components, columns = var_Df.index)
        pcaDf = pd.concat([componentsDf, df[self.y]], axis=1)
        
        self.pcaDf=pcaDf

        #Coefs
        self.coefs = pd.DataFrame((np.transpose(PCA.components_)), columns=var_Df.index, index=df.drop('status', axis=1).columns)

    def PCA_Scatterplot(self, x_='PC1', y_='PC2'):
        self.x_=x_
        self.y_=y_

        #cat=list(set(self.pcaDf[self.y]))
        cat=['Victoria','Empate','Pérdida']
        plt.figure(figsize=(8.4, 7))
        for k,i in enumerate(cat):
            color=['green', 'yellow', 'red']
            xs = self.pcaDf.loc[self.pcaDf[self.y]==i, self.x_]
            ys = self.pcaDf.loc[self.pcaDf[self.y]==i, self.y_]
            scalex = 1.0/(self.pcaDf[self.x_].max() - self.pcaDf[self.x_].min())
            scaley = 1.0/(self.pcaDf[self.y_].max() - self.pcaDf[self.y_].min())
            plt.scatter(xs * scalex, ys * scaley,s=5, label=i, color=color[k], alpha=0.7 )
            plt.legend(fontsize=14)

        plt.xlabel("{}".format(self.x_), fontsize=18)
        plt.ylabel("{}".format(self.y_), fontsize=18)
        plt.grid()

    def PCA_Biplot(self, x_='PC1', y_='PC2'):
        #los atributos x_ e y_ del Pipeline, solo pueden cambiarse con la función PCA_scatterplot

        #cat=list(set(self.pcaDf[self.y]))
        cat=['Victoria','Empate','Pérdida']
        plt.figure(figsize=(8.4, 7))
        for k,i in enumerate(cat):
            color=['green', 'yellow', 'red']
            xs = self.pcaDf.loc[self.pcaDf[self.y]==i, x_]
            ys = self.pcaDf.loc[self.pcaDf[self.y]==i, y_]
            scalex = 1.0/(self.pcaDf[x_].max() - self.pcaDf[x_].min())
            scaley = 1.0/(self.pcaDf[y_].max() - self.pcaDf[y_].min())
            plt.scatter(xs * scalex, ys * scaley,s=5, label=i, color=color[k], alpha=0.7 )
            plt.legend(fontsize=14)

        n=self.coefs.shape[0]
        for i in list(self.coefs.index):
            plt.arrow(0, 0, self.coefs.loc[i,x_], self.coefs.loc[i,y_],color = 'darkblue',alpha = 0.5)
            plt.text(self.coefs.loc[i,x_]* 1.15, self.coefs.loc[i,y_] * 1.15, i, color = 'darkblue', ha = 'center', va = 'center', fontsize=14)

        plt.xlabel("{}".format(x_), fontsize=18)
        plt.ylabel("{}".format(y_), fontsize=18)
        plt.grid()

    def Classification_Model(self, model, **kwargs):
        self.class_model = model(**kwargs)

    def Train_Classification(self, **kwargs):
        
        test_size = 0.3
        random_state = 123
        self.df_X=self.matches.drop(['home_team','home_goals', 'away_goals','away_team', self.y], axis=1)
        self.df_y=self.matches[[self.y]]

        for arg in kwargs.keys():
            if arg == 'test_size':
                test_size = kwargs[arg]
            if arg == 'random_state':
                random_state = kwargs[arg]
            if arg == 'data':
                df = kwargs[arg]
            if arg == 'y':
                y = kwargs[arg]

        from sklearn.model_selection import train_test_split
        self.X_train, self.X_test, self.y_train, self.y_test = train_test_split(self.df_X, self.df_y, test_size=test_size, random_state=random_state)

        self.class_model.fit(self.X_train, self.y_train)
        self.y_pred_train=self.class_model.predict(self.X_train)
        self.y_pred_test=self.class_model.predict(self.X_test)

    def Confusion_Matrix(self, data='test'):

        from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
        labels_=['Victoria', 'Empate', 'Pérdida']
        
        if data=='test':
            cm = confusion_matrix(self.y_test, self.y_pred_test, labels=self.class_model.classes_)
        elif data == 'train':
            cm = confusion_matrix(self.y_train, self.y_pred_train, labels=self.class_model.classes_)
        else:
            print('Escribe test o train')
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels_)
        disp.plot()
        plt.show()

    #Predicción de Mundial Qatar 2022
    def Load_World_Cup(self, qatar_matches=qatar_matches, ranking=ranking):
        qatar_matches.rename(lambda x: x.lower().strip().replace(' ', '_'), axis='columns', inplace=True)
        ranking.rename(lambda x: x.lower().strip().replace(' ', '_'), axis='columns', inplace=True)
        
        qatar_matches.drop(['year', 'date', 'host_team'], axis=1, inplace=True)
        qatar_matches.index=qatar_matches.index+1
        #qatar_matches['wcm']=1

        self.qatar_group_matches      =   qatar_matches[qatar_matches['stage']=='Group stage']
        self.qatar_round_16_matches   =   qatar_matches[qatar_matches['stage']=='Round of 16']
        self.qatar_quarter_matches    =   qatar_matches[qatar_matches['stage']=='Quarter-finals']
        self.qatar_semi_matches       =   qatar_matches[qatar_matches['stage']=='Semi-finals']
        self.qatar_third_match        =   qatar_matches[qatar_matches['stage']=='Third place']
        self.qatar_final_match        =   qatar_matches[qatar_matches['stage']=='Final']
        
        #Cambiar el index de ranking por el del equipo
        #Luego agregar el grupo al df qatar_group_matches
        ranking.index=ranking.team
        group=[]
        for i in self.qatar_group_matches.home_team:
            group.append(ranking.loc[i,'group'])
        self.qatar_group_matches['group']=group
                
        self.ranking                  =   ranking

    def Match_Prediction (self, df):
        df_=df.copy()
        col_ini=list(df.columns)
       # Agregar los atributos de cada equipo mediante su diferencia
        for i in list(df.index):
            df_.loc[i,'dif_inter_match']=self.countries.loc[df_.loc[i,'home_team'],'inter_match']-\
            self.countries.loc[df_.loc[i,'away_team'],'inter_match']

            df_.loc[i,'dif_inter_match_won']=self.countries.loc[df_.loc[i,'home_team'],'inter_match_won']-\
            self.countries.loc[df_.loc[i,'away_team'],'inter_match_won']

            df_.loc[i,'dif_inter_match_lost']=self.countries.loc[df_.loc[i,'home_team'],'inter_match_lost']-\
            self.countries.loc[df_.loc[i,'away_team'],'inter_match_lost']

            df_.loc[i,'dif_inter_match_tie']=self.countries.loc[df_.loc[i,'home_team'],'inter_match_tied']-\
            self.countries.loc[df_.loc[i,'away_team'],'inter_match_tied']

            df_.loc[i,'dif_inter_goals_sco']=self.countries.loc[df_.loc[i,'home_team'],'inter_goals_sco']-\
            self.countries.loc[df_.loc[i,'away_team'],'inter_goals_sco']

            df_.loc[i,'dif_inter_goals_con']=self.countries.loc[df_.loc[i,'home_team'],'inter_goals_con']-\
            self.countries.loc[df_.loc[i,'away_team'],'inter_goals_con']

            df_.loc[i,'dif_wc_match']=self.countries.loc[df_.loc[i,'home_team'],'wc_match']-\
            self.countries.loc[df_.loc[i,'away_team'],'wc_match']

            df_.loc[i,'dif_wc_match_won']=self.countries.loc[df_.loc[i,'home_team'],'wc_match_won']-\
            self.countries.loc[df_.loc[i,'away_team'],'wc_match_won']

            df_.loc[i,'dif_wc_match_lost']=self.countries.loc[df_.loc[i,'home_team'],'wc_match_lost']-\
            self.countries.loc[df_.loc[i,'away_team'],'wc_match_lost']

            df_.loc[i,'dif_wc_match_tied']=self.countries.loc[df_.loc[i,'home_team'],'wc_match_tied']-\
            self.countries.loc[df_.loc[i,'away_team'],'wc_match_tied']

            df_.loc[i,'dif_wc_goals_sco']=self.countries.loc[df_.loc[i,'home_team'],'wc_goals_sco']-\
            self.countries.loc[df_.loc[i,'away_team'],'wc_goals_sco']

            df_.loc[i,'dif_wc_goals_con']=self.countries.loc[df_.loc[i,'home_team'],'wc_goals_con']-\
            self.countries.loc[df_.loc[i,'away_team'],'wc_goals_con']

        df_['wcm']=1
        X=df_[list(self.X_train.columns)]
        df_=df_[col_ini]
    
        #X=df.drop(columns=['home_team', 'away_team'], axis=1)
        y_pred = self.class_model.predict(np.array(X))
        df_predicted=df_
        df_predicted['home_pred']=list(y_pred)
        df_predicted['home_pred']=df_predicted['home_pred'].astype(int).map({1:"Victoria",2:"Empate",3:"Pérdida"})
    
        self.df_predicted=df_predicted

    
    def Run_World_Cup (self):
        
        #Definición de función local para la predicción de partidos    
            ###$$$$$$$$
        def Pred_Match(df, countries=self.countries, X_train=self.X_train, class_model=self.class_model):
            df_=df.copy()
            col_ini=list(df.columns)
            # Agregar los atributos de cada equipo mediante su diferencia
            for i in list(df.index):
                df_.loc[i,'dif_inter_match']=countries.loc[df_.loc[i,'home_team'],'inter_match']-\
                countries.loc[df_.loc[i,'away_team'],'inter_match']

                df_.loc[i,'dif_inter_match_won']=countries.loc[df_.loc[i,'home_team'],'inter_match_won']-\
                countries.loc[df_.loc[i,'away_team'],'inter_match_won']

                df_.loc[i,'dif_inter_match_lost']=countries.loc[df_.loc[i,'home_team'],'inter_match_lost']-\
                countries.loc[df_.loc[i,'away_team'],'inter_match_lost']

                df_.loc[i,'dif_inter_match_tie']=countries.loc[df_.loc[i,'home_team'],'inter_match_tied']-\
                countries.loc[df_.loc[i,'away_team'],'inter_match_tied']

                df_.loc[i,'dif_inter_goals_sco']=countries.loc[df_.loc[i,'home_team'],'inter_goals_sco']-\
                countries.loc[df_.loc[i,'away_team'],'inter_goals_sco']

                df_.loc[i,'dif_inter_goals_con']=countries.loc[df_.loc[i,'home_team'],'inter_goals_con']-\
                countries.loc[df_.loc[i,'away_team'],'inter_goals_con']

                df_.loc[i,'dif_wc_match']=countries.loc[df_.loc[i,'home_team'],'wc_match']-\
                countries.loc[df_.loc[i,'away_team'],'wc_match']

                df_.loc[i,'dif_wc_match_won']=countries.loc[df_.loc[i,'home_team'],'wc_match_won']-\
                countries.loc[df_.loc[i,'away_team'],'wc_match_won']

                df_.loc[i,'dif_wc_match_lost']=countries.loc[df_.loc[i,'home_team'],'wc_match_lost']-\
                countries.loc[df_.loc[i,'away_team'],'wc_match_lost']

                df_.loc[i,'dif_wc_match_tied']=countries.loc[df_.loc[i,'home_team'],'wc_match_tied']-\
                countries.loc[df_.loc[i,'away_team'],'wc_match_tied']

                df_.loc[i,'dif_wc_goals_sco']=countries.loc[df_.loc[i,'home_team'],'wc_goals_sco']-\
                countries.loc[df_.loc[i,'away_team'],'wc_goals_sco']

                df_.loc[i,'dif_wc_goals_con']=countries.loc[df_.loc[i,'home_team'],'wc_goals_con']-\
                countries.loc[df_.loc[i,'away_team'],'wc_goals_con']

            df_['wcm']=1
            X=df_[list(X_train.columns)]
            df_=df_[col_ini]
        
            #X=df.drop(columns=['home_team', 'away_team'], axis=1)
            y_pred = class_model.predict(np.array(X))
            df_predicted=df_
            df_predicted['home_pred']=list(y_pred)
            df_predicted['home_pred']=df_predicted['home_pred'].astype(int).map({1:"Victoria",2:"Empate",3:"Pérdida"})
        
            return df_predicted

###$$$$$$$$

        #Predicción de Fase de Grupos
        self.group_predict = Pred_Match(df=self.qatar_group_matches)

        self.group_predict['away_pred']=self.group_predict.home_pred.map({'Victoria':"Pérdida",'Empate':'Empate','Pérdida':'Victoria'})

        #Fase de grupos
        home=self.group_predict[['home_team','home_pred']]
        away=self.group_predict[['away_team','away_pred']]
        home['pts']=home.home_pred.map({'Victoria':3,'Empate':1,'Pérdida':0})
        away['pts']=away.away_pred.map({'Victoria':3,'Empate':1,'Pérdida':0})

        #Group table
        gr_tbl=self.ranking.drop('fifa_ranking',axis=1)

        for i in list(gr_tbl.index):
            gr_tbl.loc[i,'pts']=home.loc[home['home_team']==i,'pts'].sum()+away.loc[away['away_team']==i,'pts'].sum()
            gr_tbl.loc[i,'fifa_ranking']=self.ranking.loc[i,'fifa_ranking']
        gr_tbl=gr_tbl.sort_values(['group','pts', 'fifa_ranking'], ascending=[True,False,True])
        gr_tbl[['pts',	'fifa_ranking']]=gr_tbl[['pts',	'fifa_ranking']].astype(int)

        self.groups_results=gr_tbl

        #Crear tabla de ganadores de Grupos
        firsts=[]
        seconds=[]
        for g in list(gr_tbl.group.unique()):
            for n,i in enumerate(list(gr_tbl.loc[gr_tbl['group']==g,'team'])):
                if n==0:
                    firsts.append(i)
                elif n==1:
                    seconds.append(i)
                else:
                    pass

        grp_wins={"group":list(ranking['group'].unique()),
        "1":firsts,
        "2":seconds}

        grp_wins=pd.DataFrame.from_dict(grp_wins)
        grp_wins['ht']=grp_wins.group.map(lambda x: str(1)+x)
        grp_wins['at']=grp_wins.group.map(lambda x: str(2)+x)
        
        self.group_winners=grp_wins[['group','1', '2']]

        #Ronda de Octavos
        round_16=self.qatar_round_16_matches.copy()
        round_16=round_16[['stage', 'home_team', 'away_team']]
        #round_16['home_t']=home_team
        #round_16['away_t']=away_team
        
        #Rellenar los espacios con los ganadores de la fase de grupos
        import itertools
        home_t=[]
        away_t=[]
        for i in list(round_16.index):
            home_t.append(grp_wins.loc[grp_wins['ht']==round_16.loc[i,'home_team']  , "1"].to_list())
            away_t.append(grp_wins.loc[grp_wins['at']==round_16.loc[i,'away_team']  , "2"].to_list())
        home_t=list(itertools.chain(*home_t))
        away_t=list(itertools.chain(*away_t))
        round_16['home_t']=home_t
        round_16['away_t']=away_t
        round_16 = round_16.rename(columns={'home_team':'ht',
                                        'away_team':"at",
                                        'home_t':"home_team",
                                        'away_t':'away_team'})
        
        #Predicción de Fase de Octavos
        round_16_predict = Pred_Match(df=round_16)

        #Tabla de resultados de Fase de Octavos
        import random
        for i in list(round_16_predict.index):
            if round_16_predict.loc[i,'home_pred']=="Victoria":
                round_16_predict.loc[i,'Winner']=round_16.loc[i,'home_team']
                
                round_16_predict.loc[i,'wm']="W"+str(i)
                
            elif round_16_predict.loc[i,'home_pred']=="Pérdida":
                round_16_predict.loc[i,'Winner']=round_16.loc[i,'away_team']
                
                round_16_predict.loc[i,'wm']="W"+str(i)
            else: #Se desempata aleatoriamente:
                if random.randint(0,1) == 0:
                    round_16_predict.loc[i,'Winner']=round_16_predict.loc[i,'away_team']
                    round_16_predict.loc[i,'wm']="W"+str(i)
                else:
                    round_16_predict.loc[i,'Winner']=round_16_predict.loc[i,'home_team']
                    round_16_predict.loc[i,'wm']="W"+str(i)
 
        self.round_16_predict=round_16_predict
                    
        #Cuartos de final
        quarter_finals_predict = self.qatar_quarter_matches.copy()
        quarter_finals_predict = quarter_finals_predict.rename(columns={'home_team':'ht','away_team':"at"})
        #quarter_finals_predict=quarter_finals.drop(['id', 'year','date','host_team'],axis=1)
        
        
        #quarter_finals_predict=self.qatar_quarter_matches.copy()
        
        for i in list(quarter_finals_predict.index):
            quarter_finals_predict.loc[i,'home_team']=(round_16_predict.loc[round_16_predict['wm']==quarter_finals_predict.loc[i,'ht'] ,'Winner'].to_string().split('    ')[1])
            quarter_finals_predict.loc[i,'away_team']=(round_16_predict.loc[round_16_predict['wm']==quarter_finals_predict.loc[i,'at'] ,'Winner'].to_string().split('    ')[1])
            quarter_finals_predict.loc[i,'wm']="W"+str(i)

        
        quarter_finals_predict = Pred_Match(df=quarter_finals_predict[['home_team','away_team']]) 
        for i in list(quarter_finals_predict.index):
            if quarter_finals_predict.loc[i,'home_pred']=="Victoria":
                quarter_finals_predict.loc[i,'Winner']=quarter_finals_predict.loc[i,'home_team']
                quarter_finals_predict.loc[i,'wm']="W"+str(i)
                
            elif quarter_finals_predict.loc[i,'home_pred']=="Pérdida":
                quarter_finals_predict.loc[i,'Winner']=quarter_finals_predict.loc[i,'away_team']
                quarter_finals_predict.loc[i,'wm']="W"+str(i)
            else: #Se desempata aleatoriamente:
                if random.randint(0,1) == 0:
                    quarter_finals_predict.loc[i,'Winner']=quarter_finals_predict.loc[i,'away_team']
                    quarter_finals_predict.loc[i,'wm']="W"+str(i)
                else:
                    quarter_finals_predict.loc[i,'Winner']=quarter_finals_predict.loc[i,'home_team']
                    quarter_finals_predict.loc[i,'wm']="W"+str(i)

        self.quarter_predict=quarter_finals_predict 
        
        #Semi finales
        semi_finals = self.qatar_semi_matches.copy()
        semi_finals=semi_finals.rename(columns={'home_team':'ht',
                                   'away_team':"at"})
        #semi_finals=semi_finals.drop(['id', 'year','date','host_team'],axis=1)

        for i in list(semi_finals.index):
            semi_finals.loc[i,'home_team']=(quarter_finals_predict.loc[quarter_finals_predict['wm']==semi_finals.loc[i,'ht'] ,'Winner'].to_string().split('    ')[1])
            semi_finals.loc[i,'away_team']=(quarter_finals_predict.loc[quarter_finals_predict['wm']==semi_finals.loc[i,'at'] ,'Winner'].to_string().split('    ')[1])
            #semi_finals.loc[i,'wm']="L"+str(i)
        
        semi_finals=Pred_Match(semi_finals[['home_team','away_team']])

        for i in list(semi_finals.index):
            if semi_finals.loc[i,'home_pred']=="Victoria":
                semi_finals.loc[i,'Winner']=semi_finals.loc[i,'home_team']
                semi_finals.loc[i,'Losser']=semi_finals.loc[i,'away_team']
                semi_finals.loc[i,'lm']="L"+str(i)
                semi_finals.loc[i,'wm']="W"+str(i)
            elif semi_finals.loc[i,'home_pred']=="Pérdida":
                semi_finals.loc[i,'Winner']=semi_finals.loc[i,'away_team']
                semi_finals.loc[i,'Losser']=semi_finals.loc[i,'home_team']
                semi_finals.loc[i,'lm']="L"+str(i)
                semi_finals.loc[i,'wm']="W"+str(i)
            else:
                if random.randint(0,1) == 0:
                    semi_finals.loc[i,'Winner']=semi_finals.loc[i,'away_team']
                    semi_finals.loc[i,'wm']="W"+str(i)
                    semi_finals.loc[i,'Losser']=semi_finals.loc[i,'home_team']
                    semi_finals.loc[i,'lm']="L"+str(i)

                else:
                    semi_finals.loc[i,'Winner']=semi_finals.loc[i,'home_team']
                    semi_finals.loc[i,'wm']="W"+str(i)
                    semi_finals.loc[i,'Losser']=semi_finals.loc[i,'away_team']
                    semi_finals.loc[i,'lm']="L"+str(i)
              
        self.semi_finals_predict = semi_finals

        #Tercer Lugar
        third_plc=self.qatar_third_match.copy()

        third_plc=third_plc.rename(columns={'home_team':'ht',
                                        'away_team':"at"})
        #third_plc=third_plc.drop(['id', 'year','date','host_team'],axis=1)
        
        for i in list(third_plc.index):
            third_plc.loc[i,'home_team']=(semi_finals.loc[semi_finals['lm']==third_plc.loc[i,'ht'] ,'Losser'].to_string().split('    ')[1])
            third_plc.loc[i,'away_team']=(semi_finals.loc[semi_finals['lm']==third_plc.loc[i,'at'] ,'Losser'].to_string().split('    ')[1])
        
        third_plc=Pred_Match(third_plc[['home_team','away_team']])

        for i in list(third_plc.index):
            if third_plc.loc[i,'home_pred']=="Win":
                third_plc.loc[i,'Winner']=third_plc.loc[i,'home_team']
                third_plc.loc[i,'wm']="W"+str(i)
                
            elif third_plc.loc[i,'home_pred']=="Lose":
                third_plc.loc[i,'Winner']=third_plc.loc[i,'away_team']
                third_plc.loc[i,'wm']="W"+str(i)
            else:
                if random.randint(0,1) == 0:
                    third_plc.loc[i,'Winner']=third_plc.loc[i,'away_team']
                    third_plc.loc[i,'wm']="W"+str(i)
                else:
                    third_plc.loc[i,'Winner']=third_plc.loc[i,'home_team']
                    third_plc.loc[i,'wm']="W"+str(i)

        self.third_place_predict=third_plc

        #Gran Final
        final=self.qatar_final_match.copy()
        final=final.rename(columns={'home_team':'ht',
                                        'away_team':"at"})
        #final=final.drop(['id', 'year','date','host_team'],axis=1)
        for i in list(final.index):
            final.loc[i,'home_team']=(semi_finals.loc[semi_finals['wm']==final.loc[i,'ht'] ,'Winner'].to_string().split('    ')[1])
            final.loc[i,'away_team']=(semi_finals.loc[semi_finals['wm']==final.loc[i,'at'] ,'Winner'].to_string().split('    ')[1])
  
        final=Pred_Match(final[['home_team','away_team']])

        for i in list(final.index):
            if final.loc[i,'home_pred']=="Victoria":
                final.loc[i,'Winner']=final.loc[i,'home_team']
                        
            elif final.loc[i,'home_pred']=="Pérdida":
                final.loc[i,'Winner']=final.loc[i,'away_team']
            else:
                if random.randint(0,1) == 0:
                    final.loc[i,'Winner']=final.loc[i,'away_team']
                else:
                    final.loc[i,'Winner']=final.loc[i,'home_team']

        self.final_predict=final
```

## Modelo de Clasificación 

El modelo de clasificación predice la 'Victoria', 'Empate' o 'Pérdida' del equipo de casa "home_team" a partir de las diferencias de atributos entre cada selección.

El método *Classification_Model* guarda el modelo de clasificación y queda abierto para que el usuario introduzca los parámetros que el modelo requiera.

Luego, el método *Train_Classification* entrena de último modelo en  *Classification_Model* ejecutado y tiene los siguientes inputs por defecto:
* test_size = 0.3
* random_state = 123
* X = df matches sin las siguientes columnas:'home_team', 'home_goals', 'away_team', 'away_goals'.
* y = columna 'status' de df matches

Por último, el método *Confusion_Matrix* depende estrictamente de la última ejecución del método *Train_Classification* y tiene como input por dejecto:
* data = 'test', para devolver la matriz de confusión con los datos de 'test'. Para solicitar la matriz de confusión con los datos de entrenamiento escribir 'train'.

A continuación, se muestran los resultados con varios modelos implementados.


```python
#Decision tree
from sklearn.tree import DecisionTreeClassifier
pipeline.Classification_Model(DecisionTreeClassifier)
pipeline.Train_Classification()
pipeline.Confusion_Matrix()
```

```python
#modelo de clasificación Gauss Naive Bayes
from sklearn.naive_bayes import GaussianNB
pipeline.Classification_Model(model=GaussianNB)
pipeline.Train_Classification()
pipeline.Confusion_Matrix()
```

```python
#Random forest
from sklearn.ensemble import RandomForestClassifier
pipeline.Classification_Model(RandomForestClassifier, max_depth=7)
pipeline.Train_Classification()
pipeline.Confusion_Matrix()
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
pipeline.Classification_Model(LogisticRegression, C=0.01, solver='liblinear')
pipeline.Train_Classification()
pipeline.Confusion_Matrix()
```
